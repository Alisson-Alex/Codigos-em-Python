```python
import matplotlib.pyplot as plt

# Dados de exemplo
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
vendas = [120, 90, 150, 80, -200, 0, 80, -150, 334, 560, 20, 599]

# Cores para as barras
cores = ['green' if venda >= 0 else 'red' for venda in vendas]

# Largura das barras
largura_barra = 0.6  # Ajuste para aumentar ou diminuir o espaço entre as barras

# Criar o gráfico de barras
plt.bar(meses, vendas, color=cores, width=largura_barra)

# Adicionar rótulos aos eixos
plt.xlabel('Mês')
plt.ylabel('Vendas (unidades)')

# Adicionar um título ao gráfico
plt.title('Vendas Mensais')

# Rotacionar os rótulos do eixo x, se necessário
plt.xticks(rotation=45)  

# Ajustar o layout para que não haja sobreposição
plt.tight_layout()  
plt.show()
```
